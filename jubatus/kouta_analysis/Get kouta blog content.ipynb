```python
from pyquery import PyQuery as pq
import re
```

```python
d = pq(url='http://yasutadayo.blog.fc2.com/blog-category-1.html')
```

```python
def remove_unnecessary_words(s):
    s = s.replace("上記の広告は１ヶ月以上更新のないブログに表示されています。 新しい記事を書く事で広告が消せます。", '')
    s = re.sub(r'●20(\d)+/(\d)+/(\d)+ : ', '', s)
    s = re.sub(r'#fc2_text.+;', '', s)
    s = re.sub(r'tag : ', '', s)
    s = re.sub(r'<!--//-->', '', s)
    tempos = [
        'ラーメン二郎 歌舞伎町店',
        'ラーメン二郎 大宮店',
        'ラーメン二郎 新宿小滝橋通り店',
        'ラーメン二郎 中山駅前店',
        'ラーメン二郎 八王子野猿街道店２',
        'ラーメン二郎 荻窪店',
        'ラーメン二郎 品川店',
        'ラーメン二郎 仙川店',
        'ラーメン二郎 環七新代田店',
        'ラーメン二郎 桜台駅前店',
        'ラーメン二郎 京成大久保店',
        'ラーメン二郎 府中店',
        'ラーメン二郎 茨城守谷店',
        'ラーメン二郎 神田神保町店',
        'ラーメン二郎 新橋店',
        'ラーメン二郎 相模大野店',
        'ラーメン二郎 松戸駅前店',
        'ラーメン二郎 千住大橋駅前店',
        'ラーメン二郎 西台駅前店',
        'ラーメン二郎 めじろ台法政大学前店',
        'ラーメン二郎 赤羽店.',
        'ラーメン二郎 池袋東口店',
        'ラーメン二郎 新小金井街道店',
        'ラーメン二郎 上野毛店',
        'ラーメン二郎 栃木街道店',
        'ラーメン二郎 仙台店',
        'ラーメン二郎 ひばりヶ丘駅前店',
        'ラーメン二郎 環七一之江店',
        'ラーメン 二郎 茨城守谷店',
        'ラーメン二郎 亀戸店'
    ]
    for tempo in tempos:
        s = s.replace(tempo, '')
    return s

def scrape_blog(baseurl='http://yasutadayo.blog.fc2.com/blog-category-1', max_items=10):
    text = ''
    for i in range(0, max_items):
        url = baseurl + '-' + str(i) + '.html'
        if i == 0:
            url = baseurl + '.html'
        d = pq(url=url)
        print('1\t%s' % (remove_unnecessary_words(d('div.entry_body').text())))
    return text
```
